```python
from sklearn import tree
import numpy as np
import pandas as pd
```

```python
skip = [i for i in range(15)]+[17]
```

```python
data0 = pd.read_table("shuffled_credit_approval_australia.dat", comment="#", sep=",", skiprows=15, header=[0])
```

```python
num_or_cat = data0.iloc[0].values
data1 = data0.drop(data0.index[0])
labels = pd.get_dummies(data1["approved"].values, drop_first=False)
labels = labels['+'].values
data = data1.drop("approved", axis="columns")
```

```python
categorical = [list(data)[i] for i in range(len(num_or_cat)) if num_or_cat[i] == 'cat']
data_with_dummies = pd.get_dummies(data, columns=categorical, drop_first=False)
categorical
```

```python
data_with_dummies = data_with_dummies.replace('?', -1)
```

```python
clf = tree.DecisionTreeClassifier(criterion = "entropy", min_samples_leaf = 37)
```

```python
train_data = data_with_dummies.iloc[:500]
train_labels = labels[:500]
train_data
```

```python
clf = clf.fit(train_data.values, train_labels)
```

```python
predictions = clf.predict(data_with_dummies.iloc[500:600])
test_labels = labels[500:600]
good = [1 for i in range(len(predictions)) if predictions[i] == test_labels[i]]
performance = len(good)/len(predictions)
performance
```
